```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Province": ["British Columbia", "Newfoundland and Labrador", "Ontario"],
    "2020": [3125.5, 262.0, 37231.5],
    "2021": [147605.0, 1385.5, 545532.0],
    "2022": [375357.0, 47773.0, 1336059.0],
    "2023": [403075.5, 56317.0, 1626254.0],
    "2024": [414863.5, 57801.0, 1700623.5]
}

df = pd.DataFrame(data)

# Transpose the DataFrame so that provinces become rows and years become columns
df = df.set_index("Province").transpose()

# Plot
plt.figure(figsize=(12, 6))
df.plot(kind='bar', stacked=True)
plt.title('Total Cases OF Covid by Province and Year')
plt.xlabel('Year')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.legend(title='Province', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
