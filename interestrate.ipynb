```python
#Import Dependancies
import pandas as pd
from pathlib import Path
```

```python
#Importing CSV files and reading the files
interest_rate_csv = Path("Resources/chartered_bank_interest.csv")
```

```python
interest_rate_df = pd.read_csv(interest_rate_csv)
```

```python
interest_rate_df
```

```python
#Remove NAN values
interest_rate_df = interest_rate_df.dropna()
interest_rate_df
```

```python
interest_rate_df = interest_rate_df.reset_index(drop=True)
```

```python
#Rename Column Names
rename_interest_df = interest_rate_df.rename(columns={"date":"Date", "V80691311":"Prime rate", "V80691333":"1 Year Conventional", "V80691334":"3 Year Conventional", "V80691335":"5 Year Conventional", 
                                             "V80691339":"1 Year Guaranteed", "V80691340":"3 Year Guaranteed", "V80691341":"5 Year Guaranteed",
                                             "V80691336":"5 Year Fixed", "V80691337":"Daily Interest Savings", "V80691338":"Non-Chequable Savings"})
rename_interest_df
```

```python
#Format Date to match the rest of the CSV Files

rename_interest_df[["Year", "Month", "Day"]] = rename_interest_df["Date"].str.split("-", expand = True)
rename_interest_df
```

```python
rename_interest_df["Year"] = rename_interest_df["Year"].astype(int)
```

```python
#Sort by certain years to match the years in other CSV Files (2005-2023)
fil_int_df = rename_interest_df.loc[(rename_interest_df["Year"] >= 2005) & (rename_interest_df["Year"]<=2023)] 
fil_int_df
```

```python
real_estate_int_df= fil_int_df.groupby(["Year"])
real_estate_int_year_df= pd.DataFrame(real_estate_int_df[["Prime rate", "1 Year Conventional", "3 Year Conventional",
                    "5 Year Conventional", "1 Year Guaranteed", "3 Year Guaranteed",
                    "5 Year Guaranteed", "5 Year Fixed", "Daily Interest Savings",
                    "Non-Chequable Savings"]].median())

real_estate_int_year_df
```

```python
#Formatting relevant Cells to Percent 
real_estate_int_year_df[["Prime rate", "1 Year Conventional", "3 Year Conventional",
                    "5 Year Conventional", "1 Year Guaranteed", "3 Year Guaranteed",
                    "5 Year Guaranteed", "5 Year Fixed", "Daily Interest Savings",
                    "Non-Chequable Savings"]] = real_estate_int_year_df[["Prime rate", "1 Year Conventional", "3 Year Conventional",
                                                                    "5 Year Conventional", "1 Year Guaranteed", "3 Year Guaranteed",
                                                                    "5 Year Guaranteed", "5 Year Fixed", "Daily Interest Savings",
                                                                    "Non-Chequable Savings"]].astype(float).map("{:.2f}%".format)

```
