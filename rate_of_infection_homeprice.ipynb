```python
# covid19 DataFrame
covid19_df = pd.read_csv("covid19_df3.csv")
```

```python
covid19_df['month'] = pd.DatetimeIndex(covid19_df['date']).month
```

```python
covid19_df['month'] = pd.to_numeric(covid19_df['month'], errors='coerce')
```

```python
# month number to month name
covid19_df['month'] = pd.to_datetime(covid19_df['month'], format='%m').dt.month_name().str.slice(stop=3)
covid19_df.head()
```

```python

# year column header to the covid19 DataFrame
covid19_df['year'] = pd.DatetimeIndex(covid19_df['date']).year
```

```python
# convert year column to numeric values
covid19_df['year'] = pd.to_numeric(covid19_df['year'], errors='coerce')
covid19_df.head()
```

```python
covid19_df.set_index('year', inplace=True)
covid19_df.head()
```

```python
# rename column headers covid19_df.
covid19_df.rename(columns={'prname': 'Province', 'pruid': 'Province_ID', 'prnameFR': 'Province_French', 'month': 'Month'},inplace=True)
# print the covid19 DataFrame
covid19_df
```

```python
covid19_df=covid19_df.dropna()
covid19_df
```

```python
merged_df = pd.merge(yearly_real_estate_df,covid19_df,how='outer',on='year')
merged_df
```

```python
merged_df.columns
merged_df.dtypes  
```

```python
merged_df= merged_df.rename(columns={"Geographical Location": "Provience","year": "Year"})
merged_df
```

```python
merged_df.dropna()
```

```python
merged_df.columns
```

```python
homeprices_df= merged_df.groupby(["Province","Year"])

homeprices_df= pd.DataFrame(homeprices_df[['Composite House',
                                           'Single Family House','One Storey House',
                                           'Two Storey House', 'Townhouse House',
                                           'Apartment House','totalcases']].median())

homeprices_df
```

```python
provinces_of_interest = ["Ontario","Newfoundland and Labrador", "British Columbia"]
provinces_of_interest
```

```python
grouped_df = homeprices_df.groupby(["Province"])
grouped_df
```

```python
# Filter the groups to include only the provinces of interest
filtered_groups = grouped_df.filter(lambda x: x.name in provinces_of_interest)
filtered_groups
```

Plotting 

```python
total_cases_by_province = filtered_groups.groupby("Province")["totalcases"].median().reset_index()
```

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(total_cases_by_province["Province"], total_cases_by_province["totalcases"], color='blue')
plt.xlabel('Province')
plt.ylabel('Total Cases')
plt.title('Total Cases by Province')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```
