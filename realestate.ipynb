```python
# using group by function to find average price for each home type by provience and year
real_estate_final_df= concat_df.groupby(["Year","Provience"])
real_estate_by_year_df= pd.DataFrame(real_estate_final_df[['Composite HPI', 'Single Family HPI',
                                                        'One Storey HPI','Two Storey HPI', 
                                                        'Townhouse HPI', 'Apartment HPI','Composite House', 
                                                        'Single Family House','One Storey House', 
                                                        'Two Storey House', 'Townhouse House', 
                                                        'Apartment House']].median())

real_estate_by_year_df
```

```python
# Creating a visulasation
real_estate_by_year_df.reset_index(inplace=True)
real_estate_by_year_df["Year"] = real_estate_by_year_df["Year"].astype(int)

# Filter the data for years 2016 to 2023
filtered_df = real_estate_by_year_df[(real_estate_by_year_df["Year"] >= 2016) & (real_estate_by_year_df["Year"] <= 2023)]

plt.figure(figsize=(12, 6))
x = filtered_df["Year"]
house_types = ["Composite House", "Single Family House", "One Storey House", "Two Storey House", "Townhouse House", "Apartment House"]
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

# Plotting the bars for each house type
for i, house_type in enumerate(house_types):
    plt.bar(x + i * 0.1, filtered_df[house_type], width=0.1, label=house_type, color=colors[i])

# Adding titles and labels
plt.title('Average Real Estate Prices by Year (2016-2023)')
plt.xlabel('Year')
plt.ylabel('Housing Price')

# Setting x-axis ticks and labels with rotation
plt.xticks(x, rotation=0)

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.savefig('Average Real Estate Price By Year.png')
plt.show()
```

```python
covid19_csv= Path("Output/covid19_df.csv")
covid19_df = pd.read_csv(covid19_csv)
covid19_df.head()
```

```python
covid19_df = covid19_df.rename(columns={"reporting_year": "Year",
                                         "Province" : "Provience" ,
                                           "totalcases" : "Total cases"})
covid19_df
```

```python
covid19_df= covid19_df.groupby(["Year","Provience"])
covid19_df= pd.DataFrame(covid19_df["Total cases"].count())
covid19_df
```

```python
real_estate_by_year_df.set_index(['Year', 'Provience'], inplace=True)
```

```python
# Now, both DataFrames have the same index names ('Year' and 'Province'), and you can proceed with the merge operation
covid_realestate_2020_2023 = pd.merge(real_estate_by_year_df, covid19_df, left_index=True, right_index=True)
```

```python
covid_cases_per_province_year = covid19_df.pivot_table(index='Year', columns='Provience', values='Total cases', aggfunc='sum')

# Plotting bar graphs for each province
for province in covid_cases_per_province_year.columns:
    plt.figure(figsize=(5, 4))
    plt.bar(covid_cases_per_province_year.index, covid_cases_per_province_year[province], width=0.5)
    plt.title(f'COVID-19 Cases in {province} per Year')
    plt.xlabel('Year')
    plt.ylabel('Total COVID-19 Cases')
    plt.xticks(covid_cases_per_province_year.index)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig('Output/Covid-19 Cases in Each Provience per Year.png')
    plt.show()
```

```python
for province in covid_realestate_2020_2023['Provience'].unique():
    # Filter data for the current province
    province_data = covid_realestate_2020_2023[covid_realestate_2020_2023['Provience'] == province]
    
    # Plot the line graph
    plt.figure(figsize=(10, 6))
    plt.plot(province_data['Year'], province_data['Composite HPI'], label='Composite HPI')
    plt.plot(province_data['Year'], province_data['Single Family HPI'], label='Single Family HPI')
    plt.plot(province_data['Year'], province_data['One Storey HPI'], label='One Storey HPI')
    plt.plot(province_data['Year'], province_data['Two Storey HPI'], label='Two Storey HPI')
    plt.plot(province_data['Year'], province_data['Townhouse HPI'], label='Townhouse HPI')
    plt.plot(province_data['Year'], province_data['Apartment HPI'], label='Apartment HPI')
    plt.title(f'Housing Price Index Increase in {province}')
    plt.xlabel('Year')
    plt.ylabel('Housing Price Index')
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.grid(True)
    plt.savefig('Output/Housing Price Increase in Each Provience.png')
    plt.show()
```
